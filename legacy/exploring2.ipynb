```python
import pandas as pd
```

```python
import numpy as np
```

```python
table = pd.read_csv("./data/table_of_appearances.csv")
```

```python
characters = pd.DataFrame({"character name": pd.Series(table["character name"]), 
                           "Appearances": pd.Series((table[:] == "Appearances").sum(axis=1)),
                           "Minor Appearances": pd.Series((table[:] == "Minor Appearances").sum(axis=1)),
                           "Mentions": pd.Series((table[:] == "Mentions").sum(axis=1))})

characters.head()
```

```python
characters["Total Appearances"] = characters.sum(axis=1)
characters.head()
```

```python
characters = characters.sort_values(by="Appearances", ascending=False, ignore_index=True)
characters
```

```python
characters.iloc[:,1:]
```

```python
table.iloc[:,1:] = table.iloc[:,1:].replace({"Appearances": 10, "Minor Appearances": 2, "Mentions": 0.5})
table.head()
```

```python
table = table.fillna(0)
table.head()
```

```python
weights = np.matmul(np.asarray(table.iloc[:,1:]), np.transpose(table.iloc[:,1:]))
weights
```

```python
weights["character name"] = table["character name"]
weights.head()
```

```python
weights.iloc[:,:-1].rename(table["character name"])
weights.head()
```
